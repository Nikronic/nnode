```python
from math import exp
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def ya(x):
    return x**2 + exp(-x**2 / 2) / (1 + x + x**3)

def dyadx(x):
    return (
        2 * x - exp(-x**2 / 2) * (1 + x + 4 * x**2 + x**4) / (1 + x + x**3)**2
    )

def d2yadx2(x):
    return (
        2 + exp(-x**2 / 2) * (1 + x * (-6 + x * (8 + x * (6 + x * (19 + x * (2 + 7 * x + x**3))))))
        / (1 + x + x**3)**3
    )
```

```python
x = np.arange(-1, 1, 0.01)
n = len(x)
y = np.zeros(n)
dy_dx = np.zeros(n)
d2y_dx2 = np.zeros(n)
for i in range(n):
    y[i] = ya(x[i])
    dy_dx[i] = dyadx(x[i])
    d2y_dx2[i] = d2yadx2(x[i])
```

```python
plt.plot(x, y, label = "$y$")
plt.plot(x, dy_dx, label = "$dy/dx$")
plt.plot(x, d2y_dx2, label = "$d^2y/dx^2$")
axes = plt.gca()
axes.set_ylim([-15, 15])
plt.xlabel("x")
plt.ylabel("$d^ky/dx^k$")
plt.title("Lagaris Problem 1 analytical solutions")
plt.legend()
```
