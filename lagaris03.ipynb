```python
from math import exp, cos, sin
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def ya(x):
    return exp(-x / 5) * sin(x)

def dyadx(x):
    return 1 / 5 * exp(-x / 5) * (5 * cos(x) - sin(x))

def d2yadx2(x):
    return -2 / 25 * exp(-x / 5) * (5 * cos(x) + 12 * sin(x))
```

```python
x = np.arange(-2, 2, 0.01)
n = len(x)
y = np.zeros(n)
dy_dx = np.zeros(n)
d2y_dx2 = np.zeros(n)
```

```python
for i in range(n):
    y[i] = ya(x[i])
    dy_dx[i] = dyadx(x[i])
    d2y_dx2[i] = d2yadx2(x[i])
```

```python
plt.plot(x, y, label = "$y$")
plt.plot(x, dy_dx, label = "$dy/dx$")
plt.plot(x, d2y_dx2, label = "$d^2y/dx^2$")
plt.xlabel("x")
plt.ylabel("$d^ky/dx^k$")
plt.legend()
plt.title("Lagaris Problem 3 analytical solutions");
```
