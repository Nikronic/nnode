```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sigma import sigma, dsigma_dz
```

```python
def ytf(A, x, N):
    return A + x*N
```

```python
def dyt_dxf(x, N, dN_dx):
    return x*dN_dx + N
```

```python
A = 0
def Gf(x, y, dy_dx):
    return x - y
```

```python
sigma_v = np.vectorize(sigma)
dsigma_dz_v = np.vectorize(dsigma_dz)
ytf_v = np.vectorize(ytf)
dyt_dxf_v = np.vectorize(dyt_dxf)
```

```python
def computeError(p, x, A, Gf):
    Gf_v = np.vectorize(Gf)
    w = p[0:H]
    u = p[H:2*H]
    v = p[2*H:3*H]
    z = np.outer(x, w) + u
    s = sigma_v(z)
    s1 = dsigma_dz_v(z)
    N = s.dot(v)
    yt = ytf_v(A, x, N)
    dN_dx = s1.dot(v*w)
    dyt_dx = dyt_dxf_v(x, N, dN_dx)
    G = Gf_v(x, yt, dyt_dx)
    E = np.sum(G**2)
    print('E =', E)
    return E
```

```python
n = 11
xt = np.linspace(0, 1, n)
H = 10
w = np.random.uniform(-1,1,H)
u = np.random.uniform(-1,1,H)
v = np.random.uniform(-1,1,H)
p = np.hstack((w, u, v))
print(w)
print(u)
print(v)
print(p)
```

```python
from scipy.optimize import minimize
res = minimize(computeError, p, args = (xt, A, Gf), method='nelder-mead', options={'disp':True,'maxfev':10000})
```

```python
print(res)
```

```python
p_new = res.x
print(p_new)
```

```python
w_new = p_new[0:H]
u_new = p_new[H:2*H]
v_new = p_new[2*H:3*H]
```

```python
z = np.outer(xt, w_new) + u_new
s = sigma_v(z)
s1 = dsigma_dz_v(z)
N = s.dot(v_new)
yt = ytf_v(A, xt, N)
dN_dx = s1.dot(v_new*w_new)
dyt_dx = dyt_dxf_v(xt, N, dN_dx)
Gf_v = np.vectorize(Gf)
G = Gf_v(xt, yt, dyt_dx)
E = np.sum(G**2)
```
